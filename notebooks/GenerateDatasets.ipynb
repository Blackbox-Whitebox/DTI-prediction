```python
from tdc.multi_pred import DTI
import os
import json
import requests
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = DTI(name = 'DAVIS')
```

```python
data.binarize(threshold = 100, order = 'descending')
balanced = data.balanced(oversample = True, seed = 42)
```

```python
train, test = train_test_split(balanced, random_state=42, train_size=.8)
```

```python
train.to_csv("davis_train.csv")
test.to_csv("davis_test.csv")
```

```python
balanced.to_csv("davis_full.csv")
```
